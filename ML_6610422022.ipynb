# Step 1. Import data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pickle
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Equ4lity/DASD6003_HW1/main/Housing.csv')
df
```

# Step 2. Check for empty data

```python
print(df.isnull().sum())
```

### Result : No missing data

```python
print(df.info)
```

# Step 3. Create model & Train model

```python
# set features
x = df[['area','bedrooms','bathrooms','parking']]

# Set target variable
y = df['price']

# Split data into training and testing sets (test size = 20%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=58)

# Create a linear regression model
lin_reg = LinearRegression()

# Train model
lin_reg.fit(x_train,y_train)

# Predic
y_pred = lin_reg.predict(x_test)
```

# Step 4. Test MSE

```python
MSE = mean_squared_error(y_test,y_pred)
print('MSE =',MSE)
```
